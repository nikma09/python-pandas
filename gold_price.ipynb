```python
import pandas as pd
```

```python
df = pd.read_csv('Gold Price.csv')
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'],format= '%d-%m-%Y')
```

```python
df.head(25)
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()
```

```python
monthly_avg = df.groupby(['Year', 'Month'])['Price'].mean().reset_index()
```

```python
monthly_avg.head(20)
```

```python
df['Month_Num'] = df['Date'].dt.month
monthly_avg = df.groupby(['Year', 'Month_Num', 'Month'])['Price'].mean().reset_index()
monthly_avg = monthly_avg.sort_values(['Year', 'Month_Num'])
monthly_avg = monthly_avg[['Year', 'Month', 'Price']]
```

```python
monthly_avg.head(12)
```

```python
monthly_avg.to_csv('Monthly_Gold_Price_2020_2025.csv', index=False)
```
