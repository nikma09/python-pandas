```python
import pandas as pd
```

```python
def my_sq(x):
    return x**2
```

```python
# apply basic function 
df = pd.DataFrame({"a":[10,20,30],"b":[20,30,40]})
print(df['a']**2)
```

```python
df
```

```python
# apply over series
df['a'].apply(my_sq)
```

```python
# apply over dataframe
# we pass axis=0 or axis='index' to apply method column wise axis=1 or axis ='column' to apply method row wise
def print_me(x):
    return(x)
# column wise operation, axis = 0
df.apply(print_me, axis=0)
# column wise operation, axis = 'index'
df.apply(print_me, axis='index')
```

```python
# row wise operation, axis = 1
df.apply(print_me, axis=1)
# row wise operation, axis = 'columns'
df.apply(print_me, axis='columns')
```

```python
def avg(col):
    x = col[0]
    y = col[1]
    z = col[2] 
    return (x+y+z)/3
df.apply(avg,axis=0)
```

```python
def avg(col):
    x = col[0]
    y = col[1]
    return (x+y)/2
df.apply(avg,axis=1)
```

# vectorize with numpy

```python
import numpy as np
```

```python
def avg2(x,y):
    if x == 20:
        return np.nan
    else:
        return (x+y)/2
```

```python
# after vectorizing function can accepct vector as argument
average = np.vectorize(avg2)
```

```python
print(average(df['a'],df['b']))
```
