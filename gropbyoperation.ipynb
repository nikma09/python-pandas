# aggrigation function
## custom function

```python
def my_mean(val):
    n=len(val)
    return val.sum()/n
```

```python
df.groupby('year')['lifeExp'].agg(my_mean)
```

```python
diff_val = df['lifeExp'].mean()
diff_val
```

```python
def diffmean(val,diff):
    n = len(val)
    Sum = 0
    for value in val:
        Sum += value
    mean = Sum/n
    return mean - diff     
```

```python
mean_diff = df.groupby('year')['lifeExp'].agg(diffmean,diff=diff_val)
```

```python
mean_diff
```

# Transform

## Z score Example
### formula = x-µ/σ x= data
### point in dataframe 
### µ= average of dataframe= sum of data points/no. of data points
### σ= standerd SD = √[ Σ(x − µ)² / n ]


```python
def my_zval(x):
    return (x-x.mean())/x.std()
```

```python
# calculating grouped data zscors
z_value = df.groupby('year')['lifeExp'].transform(my_zval)
```

```python
z_value
```
